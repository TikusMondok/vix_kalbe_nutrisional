```python
import pandas as pd

```

```python
df_customer = pd.read_csv('./case_study/Case Study - Customer.csv', delimiter=';')
df_store = pd.read_csv('./case_study/Case Study - Store.csv', delimiter=';')
df_product = pd.read_csv('./case_study/Case Study - Product.csv', delimiter=';')
df_trx = pd.read_csv('./case_study/Case Study - Transaction.csv', delimiter=';')
```

CLEANSING DATA

```python
#cleansing data customer
df_customer['Income'] = df_customer['Income'].replace('[,]', '.', regex=True).astype('float')
```

```python
df_customer.sample(5)
```

```python
#cleansing data store
df_store['Latitude'] = df_store['Latitude'].replace('[,]', '.', regex=True).astype('float')
df_store['Longitude'] = df_store['Longitude'].replace('[,]', '.', regex=True).astype('float')
```

```python
#cleasning data transaction
df_trx['Date'] = pd.to_datetime(df_trx['Date'], format="%d/%m/%Y")
```

```python
df_trx.sample(5)
# df_trx.info()
```

MENGGABUNGKAN DATA

```python
df_merge = pd.merge(df_trx, df_customer, on=['CustomerID'])
df_merge = pd.merge(df_merge, df_product.drop(columns=['Price']), on=['ProductID'])
df_merge = pd.merge(df_merge, df_store, on=['StoreID'])
```

```python
df_merge.sample(5)
```

```python
df_merge.to_csv('./case_study/dataset_merge.csv', index=False)
```
